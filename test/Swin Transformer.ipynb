```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
```

```python
relative_coords_h = torch.arange(-(3 - 1), 3, dtype=torch.float32)
relative_coords_w = torch.arange(-(3 - 1), 3, dtype=torch.float32)
relative_coords_table = torch.stack(torch.meshgrid([relative_coords_h,
                            relative_coords_w])).permute(1, 2, 0).contiguous().unsqueeze(0)
print(relative_coords_h, relative_coords_w)
print(relative_coords_table.shape)
```

```python
relative_coords_table[:, :, :, 0] /=(3-1)
relative_coords_table[:, :, :, 1] /=(3-1)
print(relative_coords_table)
```

```python
relative_coords_table[:, :, :, 0] *= 8
relative_coords_table = torch.sign(relative_coords_table)* torch.log2(
    torch.abs(relative_coords_table) + 1.0) / np.log2(8)
print(relative_coords_table)
```
